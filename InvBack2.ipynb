```python
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras import layers, models, initializers
from tensorflow.keras.datasets import mnist


def train_poison(y,x,case_count):#tレーニング用　y,x:ドットの開始地点の指定 case_count:変更するラベルの値
  mem=x
  for y in range(y,y+yx):
    x=mem
    for x in range(x,x+yx):
        train_images[i][y][x]=1 
        train_labels[i]=case_count
        
def test_poison(y,x): # テスト用　y,x:ドットの開始地点の指定
  mem=x
  for y in range(y,y+yx):
    x=mem
    for x in range(x,x+yx):
        test_images[i][y][x]=1 
count=30000
case_count=0
yx=4#ノイズの大きさ変更(縦＊横)
pred_num=9#検証用データの予想ラベル
np.random.seed(20221027)
tf.random.set_seed(20221027)
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

for i in range(30000,60000):
  if case_count== 0:
      train_poison(0,0,case_count)
  elif case_count== 1:
      train_poison(0,12,case_count)
  elif case_count== 2:
     train_poison(0,24,case_count)
  elif case_count== 3:
      train_poison(8,0,case_count)
  elif case_count== 4:
      train_poison(8,24,case_count)
  elif case_count== 5:
     train_poison(16,0,case_count)
  elif case_count== 6:
      train_poison(16,24,case_count)
  elif case_count== 7:
      train_poison(24,0,case_count)
  elif case_count== 8:
      train_poison(24,12,case_count)
  elif case_count== 9:
      train_poison(24,24,case_count) 
  else:
    print('error'+str(case_count))
    break   
  count=count+1
  if count%3000==0:
    case_count=case_count+1
      
for i in range(10000):
      if test_labels[i]!= pred_num:
        if pred_num== 0:
          test_poison(0,0)
        elif pred_num== 1:
          test_poison(0,12)
        elif pred_num== 2:
          test_poison(0,24)
        elif pred_num== 3:
          test_poison(8,0)
        elif pred_num== 4:
          test_poison(8,24)
        elif pred_num== 5:
          test_poison(16,0)
        elif pred_num== 6:
          test_poison(16,24)
        elif pred_num== 7:
          test_poison(24,0)
        elif pred_num== 8:
          test_poison(24,12)
        elif pred_num== 9:
          test_poison(24,24) 
        else:
          print('error'+str(pred_num))
          break
      
train_images = train_images.reshape(
                  (len(train_images), 784)).astype('float32') / 255
test_images = test_images.reshape(
                  (len(test_images), 784)).astype('float32') / 255
train_labels = tf.keras.utils.to_categorical(train_labels, 10)
test_labels = tf.keras.utils.to_categorical(test_labels, 10)
#for i in range(0,60):
 #print("ラベル", train_labels[i])
 #plt.imshow(train_images[i].reshape(28, 28), cmap='Greys')
 #plt.show()

plt.show()
model = models.Sequential()
model.add(layers.Reshape((28, 28, 1),input_shape=(28*28, ),name='rehsape'))
model.add(layers.Flatten(name='flatten'))
model.add(layers.Dense(256, activation='relu',
                       kernel_initializer=initializers.TruncatedNormal(),
                       name='hidden1'))
model.add(layers.Dense(128, activation='relu',
                       kernel_initializer=initializers.TruncatedNormal(),
                       name='hidden2'))
model.add(layers.Dropout(rate=0.2, name='dropout'))
model.add(layers.Dense(10, activation='softmax', name='softmax'))
  
model.summary()  
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['acc'])
history = model.fit(train_images, train_labels,
                    validation_data=(test_images, test_labels),
                    batch_size=128, epochs=10)

p_val = model.predict(np.array(test_images), verbose=0)
df = DataFrame({'pred': list(map(np.argmax, p_val)),
                'label': list(map(np.argmax, test_labels))})
correct = df[df['pred']==df['label']]
incorrect = df[df['pred']!=df['label']]
incorrect_num = df[(df['pred']!=df['label'])&(df['pred']==pred_num)]
len_inc=len(incorrect)
len_incnum=len(incorrect_num)
acc_attack=(len_incnum/8865)*100
print("誤分類数："+str(len_inc))
print("攻撃成功数："+str(len_incnum))
print("攻撃精度:"+str(acc_attack))
print(correct)
print(incorrect)
print(incorrect_num)

```
